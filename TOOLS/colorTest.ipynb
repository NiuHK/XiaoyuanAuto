```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import product
def hex_to_rgb(hex_color):
    # 移除16进制颜色中的'#'符号（如果有）
    hex_color = hex_color.lstrip('#')

    # 按照每两位分割16进制颜色，转换为十进制的RGB值
    rgb = [int(hex_color[i:i+2], 16) for i in (0, 2, 4)]
    
    return rgb

# 输入目标颜色和距离阈值
# target_color = np.array([247, 226, 237])  # RGB format
target_color=hex_to_rgb('#fce1ee')
threshold = 50

# 定义欧几里得距离计算函数
def euclidean_distance(color1, color2):
    return np.sqrt(np.sum((color1 - color2) ** 2))

# 生成所有可能的颜色（RGB每个通道值为0到255）
all_colors = np.array(list(product(range(256), repeat=3)))

# 计算所有颜色到目标颜色的欧几里得距离
distances = np.linalg.norm(all_colors - target_color, axis=1)

# 筛选出距离小于x的颜色
filtered_colors = all_colors[distances < threshold]

# 可视化结果
fig, ax = plt.subplots(figsize=(10, 6))
ax.imshow([filtered_colors / 255.0], aspect="auto")
ax.set_title(f"Colors within distance {threshold} from {target_color}")
ax.axis("off")
plt.show()
```
